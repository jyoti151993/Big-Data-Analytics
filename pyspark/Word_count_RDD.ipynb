```python
import findspark
findspark.init("/usr/local/spark")
import pyspark
sc=pyspark.SparkContext(appName="word count")
```

```python
lines=sc.textFile("/home/hduser/Downloads/war_and_peace.txt",2)
```

```python
lines.count()
```

```python
lines.take(5)
```

```python
lines.getNumPartitions()
```

```python
def count_in_a_partition(iterator):
    yields sum(1 for _ in iterator)
lines.mappartition(count_in_a_partition).collect()    
```

```python
noNullLines=lines.filter(lambda line:len(line)>0)
```

```python
noNullLines.count()
```

```python
noNullLines.take(5)
```

```python
word=noNullLines.flatMap(lambda x:x.split())
```

```python
word.count()
```

```python
word.take(5)
```

```python
upper=word.map(lambda x:x.upper())
```

```python
upper.take(5)
```

```python
upper.collect()
```

```python
pairedOnes=upper.map(lambda uv:(uv,1))
```

```python
pairedOnes.take(5)
```

```python
wordCount=pairedOnes.reduceByKey(lambda prev,next:(prev+next))
```

```python
wordCount.take(20)
```

```python
for word in wordCount.take(5):
    print("*******",word)
```

```python
wordCount.count()
```
